```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/saro/Desktop/dvdrentaldl/dvd rental final csv.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
customer_rental = df.groupby('customer_id').size().sort_values(ascending=False)

plt.figure(figsize=(20,6))
customer_rental.head(80).plot(kind='bar',color='green',alpha=0.7)

# Adding title and labels
plt.title('Number of Rentals per Customer')
plt.xlabel('Customer ID')
plt.ylabel('Number of Rentals')

# Show the plot
plt.show()
```

```python
rentals_per_day = df.resample('D', on='rental_date').size()

# Plotting the number of rentals over time
plt.figure(figsize=(10, 6))
plt.plot(rentals_per_day.index, rentals_per_day.values, color='green', alpha=0.7)

# Adding title and labels
plt.title('Number of Rentals Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Rentals')

# Show the plot
plt.show()
```

```python
# 1. Rental Rates Distribution
plt.figure(figsize=(10, 6))
rental_rate_counts = df['rental_rate'].value_counts().sort_index()
rental_rate_counts.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.hist(df['rental_rate'], color='skyblue', edgecolor='black')
plt.title('Distribution of Rental Rates')
plt.xlabel('Rental Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Film Length Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['length'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Film Length')
plt.xlabel('Film Length (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Top Film Categories (Bar Plot)
category_counts = df['category'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Top Film Categories')
plt.xlabel('Category')
plt.ylabel('Number of Films')
plt.xticks(rotation=45)
plt.show()
```
